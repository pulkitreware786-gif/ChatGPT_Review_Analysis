### Sentiment Analysis

Now we will analyze the sentiment of the reviews using `TextBlob`. We will create `sentiment_polarity` (a number indicating sentiment) and `sentiment_label` (Positive, Negative, Neutral) for each review.

```python
from textblob import TextBlob

# Function to get sentiment polarity
def get_sentiment_polarity(text):
    if isinstance(text, str):
        return TextBlob(text).sentiment.polarity
    return None

# Function to get sentiment label based on polarity
def get_sentiment_label(polarity):
    if polarity is None:
        return 'Neutral'
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the 'Review' column
df['sentiment_polarity'] = df['Review'].apply(get_sentiment_polarity)
df['sentiment_label'] = df['sentiment_polarity'].apply(get_sentiment_label)

# Display the DataFrame with new sentiment columns and basic sentiment distribution
display(df[['Review', 'sentiment_polarity', 'sentiment_label']].head())
display(df['sentiment_label'].value_counts())
```

### Issue Identification: Common Words in Negative Reviews

Now we will find the most common words in negative reviews, so we can understand what problems users are facing.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import re
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for negative reviews
negative_reviews_df = df[df['sentiment_label'] == 'Negative']

# Combine all negative reviews into a single string
all_negative_reviews_text = ' '.join(negative_reviews_df['Review'].dropna().tolist())

# Preprocess the text
# Convert to lowercase
text = all_negative_reviews_text.lower()
# Remove punctuation and numbers
text = re.sub(r'[^a-z\s]', '', text)
# Tokenize words
words = word_tokenize(text)
# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words and len(word) > 1]

# Get the most common words
word_counts = Counter(filtered_words)
most_common_words = word_counts.most_common(20)

# Display the most common words
print("Top 20 most common words in negative reviews:")
for word, count in most_common_words:
    print(f"{word}: {count}")

# Prepare data for visualization
words_df = pd.DataFrame(most_common_words, columns=['Word', 'Count'])

# Visualize the most common words
fig = plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Word', data=words_df, palette='viridis')
plt.title('Top 20 Most Common Words in Negative Reviews')
plt.xlabel('Count')
plt.ylabel('Word')
plt.show()
```

### Time-Series Analysis: Sentiment Evolution Over Time

We will now see how the average sentiment of reviews changes over time.

```python
# Convert 'Review Date' to datetime objects
df['Review Date'] = pd.to_datetime(df['Review Date'])

# Sort by date to ensure correct time series order
df = df.sort_values(by='Review Date')

# Aggregate sentiment polarity by day
daily_sentiment = df.groupby(df['Review Date'].dt.date)['sentiment_polarity'].mean().reset_index()
daily_sentiment['Review Date'] = pd.to_datetime(daily_sentiment['Review Date'])

display(daily_sentiment.head())
```

### Visualize Sentiment Evolution

Now we will display the daily average sentiment in a line plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize sentiment evolution over time
fig = plt.figure(figsize=(14, 7))
sns.lineplot(x='Review Date', y='sentiment_polarity', data=daily_sentiment)
plt.title('Average Sentiment Polarity Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sentiment Polarity')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
